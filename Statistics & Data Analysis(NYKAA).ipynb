# Web Scraping Nykaa

# Step 1 : Install / Import the necessary libraries

```python
# import web scrapping and data manipulation libraries


import requests
from bs4 import BeautifulSoup as soup
import numpy as np
import pandas as pd
```

# Step 2 : Choose the data you need to extract

```python
# scrap the mamaearth product

p_url = 'https://www.nykaa.com/skin/masks/masks-peels/c/8400'
p_content = requests.get(p_url)
```

```python
# to find the status code

p_content.status_code
```

# Step 3 : Send HTTP request to the URL of the page you want to scrape

Now that we have got the HTML content of the page, we need to parse the data and store it in soup variable to structure it into a different format which will help in data extraction.

```python
# to print the things in url in the form of html

data = soup(p_content.text,'html.parser')
data
```

```python
container=data.find_all("div",{"class":'css-d5z3ro'})
```

```python
container
```

```python
con= container[0]
print(soup.prettify(con))
```

```python
sel_class = 'css-xrzmfa'
name_title = data.find_all('div', class_=sel_class)
```

```python
len(name_title)
```

# Step 4 : Inspect the page and write codes for extraction

# Product name

```python
#to get the product names

for name in name_title:
    main_title = name.text.strip()
    print(main_title)
```

# Price

```python
pr_class = 'css-111z9ua'
price_rs = data.find_all('span', class_=pr_class)
```

```python
#to find number of prices

len(price_rs)
```

```python
#to print all prices

for name in price_rs:
    main_title = name.text.strip()
    print(main_title)
```

# Offer

```python
offer_class = 'css-cjd9an'
offer_rs = data.find_all('span', class_=offer_class)
```

```python
len(offer_rs)
```

```python
# to get offers

for name in offer_rs:
    main_title = name.text.strip()
    print(main_title)
```

```python
class1 = 'css-qlopj4'
doc1 = data.find_all('a', class_ = class1)
```

```python
# as there are 20 prodcuts in the link

len(doc1)
```

```python
# taking 1st element out of 20

first_name = doc1[0]
print(first_name)
```

```python
# printing details of one product

product_name = first_name.find('div', class_ = 'css-xrzmfa').text.strip()
price = first_name.find('span', class_ = 'css-111z9ua').text.replace('₹','').strip()
rating = first_name.find('span', class_ = 'css-1qbvrhp').text.strip()
```

```python
print(product_name, price, rating)
```

```python
main_list = []

for doc in doc1:
    temp_dict={}
    temp_dict['product_name'] = doc.find('div', class_ = 'css-xrzmfa').text.strip()
    temp_dict['price'] = doc.find('span', class_ = 'css-111z9ua').text.replace('₹','').strip()
    temp_dict['rating'] = doc.find('span', class_ = 'css-1qbvrhp').text.strip()
    main_list.append(temp_dict)
```

```python
# to create it as a dataframe

main_df = pd.DataFrame(main_list)
```

```python
# to print the dataframe

main_df
```
