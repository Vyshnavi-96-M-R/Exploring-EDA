# Outlier Analysis

```python
fig, axs = plt.subplots(2,3, figsize = (10,5))
plt1 = sns.boxplot(house_df['price'], ax = axs[0,0])
plt2 = sns.boxplot(house_df['bedrooms'], ax = axs[0,1])
plt3 = sns.boxplot(house_df['bathrooms'], ax = axs[0,2])
plt1 = sns.boxplot(house_df['floors'], ax = axs[1,0])
plt2 = sns.boxplot(house_df['view'], ax = axs[1,1])
plt3 = sns.boxplot(house_df['condition'], ax = axs[1,2])

plt.tight_layout()
```

# Outlier Treatment

Price is considerable as outliers.We can drop the outliers as we have sufficient data.

```python
# outlier treatment for price

plt.boxplot(house_df.price)
Q1 = house_df.price.quantile(0.25)
Q3 = house_df.price.quantile(0.75)
IQR = Q3 - Q1
house_df = house_df[(house_df.price >= Q1 - 1.5*IQR) & (house_df.price <= Q3 + 1.5*IQR)]
```

# Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# We specify this so that the train and test data set always have the same rows, respectively

np.random.seed(0)
house_df_train, house_df_test = train_test_split(house_df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to some of the columns 

num_vars = ['price', 'bedrooms', 'bathrooms', 'floors', 'view']

house_df_train[num_vars] = scaler.fit_transform(house_df_train[num_vars])
```

# Dividing into X and Y sets for the model building

```python
y_train = house_df_train.pop('price')
X_train = house_df_train
```

# Model Building

This time, we will be using the LinearRegression function from SciKit Learn for its compatibility with RFE (which is a utility from sklearn)

RFE

Recursive feature elimination

```python
# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
# Running RFE with the output number of the variable equal to 10

lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
print(f'coef:{lm.coef_}')
print(f'intercept:{lm.intercept_}')
```

```python
# Displaying accuracy for training set

y_pred=lm.predict(X_train)
r_squared=r2_score(y_train,y_pred)
print(f'Training Score:{r_squared}')
```
