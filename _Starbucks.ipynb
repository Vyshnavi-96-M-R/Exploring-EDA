# Step 3 : EDA and Visualization

```python
# Analysing the csv Beverage.csv
```

```python
# printing the Beverage category

starbuck_df['Beverage_category'].unique()
```

```python
# plot beverage category

starbuck_df.calories =starbuck_df[['Beverage_category','Calories']]
starbuck_df.calories
```

```python
sns.countplot(y='Beverage_category',data=starbuck_df.calories)
```

```python
# Create a DataFrame to print number of drinks by Beverage_category
# assign "BeverageCategory" as index 


starbuck_df_beverage = pd.DataFrame(starbuck_df['Beverage_category'].value_counts())
starbuck_df_beverage['BeverageCategory'] = starbuck_df_beverage.index
starbuck_df_beverage.columns = ['NumberofDrinks', 'BeverageCategory']
starbuck_df_beverage = starbuck_df_beverage.reset_index().drop('index', axis=1)
starbuck_df_beverage.head()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(y='Beverage_category',data=starbuck_df)
plt.title("\nPriority Drinks\n\n")
```

```python
# which Starbucks drink has the highes calories Beverage_category

starbuck_df.calories.sort_values("Calories",ascending=False)
```

```python
# data visualization to see the which Starbucks drink has the highest calories

plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y="Calories", data=starbuck_df)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Calories")
plt.show()
```

```python
# which Starbucks drink has the highest sugars

starbuck_df.sort_values(" Sugars (g)",ascending=False)
```

```python
# data visualization to see the which Starbucks has the highest calories


plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y=" Sugars (g)", data=starbuck_df)
plt.xticks(rotation=40, ha='right')
plt.title("Starbucks Beverage Sugar(g)")
plt.show()
```

```python
#plot a historgram of calories data

calories =starbuck_df["Calories"]
plt.hist(calories, bins=9, edgecolor = "black", color = "orange")
plt.title("Calories in Starbucks Menu Items")
plt.xlabel("Calories in kcal")
plt.ylabel("Count")
```

```python
#plot a historgram of total fat(g) data

fat =starbuck_df[" Total Fat (g)"]
plt.hist(fat, bins=9, edgecolor = "black", color = "green")
plt.title("Fat in Starbucks Menu Items")
plt.xlabel("Fat in kcal")
plt.ylabel("Count")
```

```python
#plot a historgram of sugar data

fat =starbuck_df[" Sugars (g)"]
plt.hist(fat, bins=9, edgecolor = "black", color = "yellow")
plt.title("Sugars in Starbucks Menu Items")
plt.xlabel("Sugar in g")
plt.ylabel("Count")
```

```python
# Drinks Analysis
# plot histogram of fat

fat =stardrink_df["Fat (g)"]
plt.hist(fat, bins=9, edgecolor = "black", color = "Blue")
plt.title("Fat in Starbuck_drink") 
plt.xlabel("Fat")
plt.ylabel("Count")
```

```python
# plot histogram of sodium

fat =stardrink_df["Sodium"]
plt.hist(fat, bins=9, edgecolor = "black", color = "pink")
plt.title("Sodium in Starbuck_drink") 
plt.xlabel("Sodium")
plt.ylabel("Count")
```
